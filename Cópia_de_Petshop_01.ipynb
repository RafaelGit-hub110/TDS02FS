```python
class PetShop:
  def __init__(self, nome, raca, peso, idade):
    self.nome = nome
    self.raca = raca
    self.peso = peso
    self.idade = idade

  def __str__(self):
    return f"Nome: {self.nome}, raça: {self.raca}, peso: {self.peso}, idade: {self.idade}"
```

```python
class Servico:
  def __init__(self, tosa, banho):
    self.tosa = tosa
    self.banho = banho

  def __str__(self):
    return f"Tosa: {self.tosa}, banho: {self.banho}"
```
